# Задание 1. Множества

```python
hw_set = {1,3,8,9,10,15,16,16}
```

```python
##############################################   Задание 1.1   #################################################################

### Как вы думаете, чему равна длина данного множества? Выведите его длину с помощью функции len
print(len(hw_set))
```

```python
##############################################   Задание 1.2   #################################################################

### Создайте пустое множество hw_set_sq и с помощью цикла запишите в него возведенные в квадрат элементы множества hw_set

hw_set_sq = set()
for i in hw_set:
    hw_set_sq.add(i**2)

print(hw_set_sq)
```

```python
##############################################   Задание 1.3   #################################################################

### Добавьте к hw_set элементы 'my_string' и 'your_string'
### Повторите действия из 1.2, но добавьте проверку на тип данных 

hw_set_sq = set()
hw_set.add('my_string')
hw_set.add('your_string')

for i in hw_set:
    if type(i) != str:
        hw_set_sq.add(i**2)

print(hw_set_sq)
print(hw_set)
```

```python
##############################################   Задание 1.4   #################################################################

### Создайте копию множества hw_set (назовите его hw_set_copy) 
### Используйте цикла (счетчик пробегает по hw_set_copy) и условия для модификации множества hw_set
### Добавьте к множеству hw_set квадраты всех цифровых элементов и первые символы всех строк

hw_set = {1,3,8,9,10,15,16,16,'my_string','your_string'}
hw_set_copy = hw_set.copy()

for i in hw_set_copy:
    if type(i) != str:
        hw_set.add(i**2)
    else:
        hw_set.add(i[:1])

print(hw_set)
```

```python
##############################################   Задание 1.5   #################################################################

### Создайте копию множества hw_set (назовите его hw_set_copy)
### Используйте цикла (счетчик пробегает по hw_set_copy) и условия для модификации множества hw_set
### Удалите из множества hw_set все элементы-строки

hw_set = {1,3,8,9,10,15,16,16,'my_string','your_string'}
hw_set_copy = hw_set.copy()

for i in hw_set_copy:
    if type(i) == str:
        hw_set.discard(i)
    

print(hw_set)
```

# Задание 2. Словари

```python
employees_0 = {'Bob':'Dev','Albert':'Manager','Victor':'Analyst','Anna':'Analyst'}
```

```python
##############################################   Задание 2.1   #################################################################

### Выведите имена всех сотрудников, кто не является менеджером. Используйте цикл for с условием внутри.
for worker in employees_0:
    if employees_0[worker] != "Manager":
        print(worker)
```

```python
##############################################   Задание 2.2   #################################################################

### Посчитайте количество сотрудников-аналитиков. Используйте цикл for с условием внутри.
count = 0
for worker in employees_0:
    if employees_0[worker] == "Analyst":
        count += 1

print(count)        
```

```python
##############################################   Задание 2.3   #################################################################

### Посчитайте количество сотрудников, которые или являются аналитиками, или их зовут Боб. 
### Используйте цикл for с условием внутри.
count = 0
for worker in employees_0:
    if employees_0[worker] == "Analyst" or worker == "Bob" :
        count += 1

print(count)  
```

```python
##############################################   Задание 2.4   #################################################################

##### Изменим немного структуру словаря

employees_1 = {'name':['Bob','Albert','Victor', 'Anna'], 'job':['Dev','Manager','Analyst','Analyst']}

### Выведите профессию первого сотрудника в словаре
print(employees_1['job'][0])
```

```python
##############################################   Задание 2.5   #################################################################

### Выведите количество сотрудников-аналитиков в словаре. Используйте цикл с условием.
count = 0
for j in employees_1['job']:
    if j == "Analyst":
        count += 1

print(count)
```

# Задание 3. Цикл while

```python
##############################################   Задание 3.1   #################################################################

### С помощью цикла while записывайте в список lst_sum все элементы списка lst_init, пока сумма элементов в lst_sum не достигнет 100
### Вам потребуются методы списков append и pop

lst_init = [4, 17, 22, 10, 8, 15, 8, 9, 2, 32, 13, 11, 12, 10, 9, 12, 22, 23, 13, 11, 10, 7, 42, 43, 14, 15, 16]
lst_sum = []
while sum(lst_sum) < 100:
    n = lst_init.pop()
    lst_sum.append(n)
    print("Текущая сумма" ,sum(lst_sum))

print(sum(lst_sum))
```

```python
##############################################   Задание 3.2   #################################################################

### Создайте переменную bin_var, котор является параметром и может принимать значения 1 или 0. 
### Если она равняется другому, то выводится ошибка "Переменная bin_var должна равняться или 1, или 0!"
### Если она равняется нулю, то выполняется действие из задания 3.1
### Если она равняется единице, то действие из задания 3.1 выполняется не с конца списка lst_init, а с начала
### Для этого вам потребуется метод списка reverse(). Не забудьте пересоздать развернутый наоборот список внутри условия

lst_init = [4, 17, 22, 10, 8, 15, 8, 9, 2, 32, 13, 11, 12, 10, 9, 12, 22, 23, 13, 11, 10, 7, 42, 43, 14, 15, 16]
lst_sum = []

bin_var = int(input('Введи 1 или 0'))

if bin_var != 1 and bin_var != 0:
    print("Переменная bin_var должна равняться или 1, или 0!")
elif bin_var == 0:
    while sum(lst_sum) < 100:
        n = lst_init.pop()
        lst_sum.append(n)
        print("Текущая сумма" ,sum(lst_sum))
else:
    lst_init.reverse()
    while sum(lst_sum) < 100:
        n = lst_init.pop()
        lst_sum.append(n)
        print("Текущая сумма" ,sum(lst_sum))

print(sum(lst_sum))
```
