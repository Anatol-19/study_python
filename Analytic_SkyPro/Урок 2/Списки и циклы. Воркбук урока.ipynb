# 7. Списки внутри списков

```python
nested_list = [2, 'Mr White', [1,2,3], (1,2,3), 3, 'California', 14.09, 15.0, 15, False]
```

```python
for i in nested_list:
    print(i)
    print(type(i))
```

Выведем только integer

```python
for i in nested_list:
    if type(i) == int:
        print(i)
```

Создадим лист, в который сохраним все integer из nested_list

```python
int_in_nested_list = []

for i in nested_list:
    if type(i) == int:
        int_in_nested_list.append(i)
        
print(int_in_nested_list)
```

Создадим лист, в который сохраним все одиночные значения из nested_list (векторные типы данных разложим)

```python
elements_nested_list = []

for i in nested_list:
    if (type(i)==list) | (type(i)==tuple):
        for j in i:
            elements_nested_list.append(j)
    else:
        elements_nested_list.append(i)
        
print(elements_nested_list)
```

# 9. Задача

**Налогообложение цен подчиняется следующей схеме:**
- Если цена меньше 1000 рублей, то НДС составляет 2.5%
- Если цена меньше 5000 рублей, то НДС составляет 3.1%
- Если цена меньше 10000 рублей, то НДС составляет 3.5%
- Если цена больше 10000 рублей, то НДС составляет 4%


Рассчитаем цены с учетом налогообложения и запишем в отдельный список

```python
prices = [59, 79, 999, 1399, 9999, 10679, 11200, 12324, 856, 1097, 1099, 2500, 665]
```

```python
print(len(prices))
```

```python
indexed_prices = []

for i in prices:
    if i <= 1000:
        new_price = i * 1.025
    elif i <= 5000:
        new_price = i * 1.031
    elif i <= 10000:
        new_price = i * 1.035
    else:
         new_price = i * 1.04
    
    indexed_prices.append(new_price)
    
print(indexed_prices)
```

Используем функцию ROUND, чтобы округлить результаты до 2 знаков после запятой

```python
##### https://pythonru.com/uroki/funkcija-round-dlja-nachinajushhih
```

```python
indexed_prices = []

for i in prices:
    if i <= 1000:
        new_price = i * 1.025
    elif i <= 5000:
        new_price = i * 1.031
    elif i <= 10000:
        new_price = i * 1.035
    else:
        new_price = i * 1.04
    
    indexed_prices.append(round(new_price,2))
    
print(indexed_prices)
```
